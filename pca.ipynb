# PCA 

Principal Component Analysis is a descomposition algorithm use in dimensionality reducction problems which mean, that it's used to reduce the number of variables in the data. 

The outcome data it's not a selection of features, there are new number *X* **virtual  features** that are a mixed of the old  features and contain the relevan information. 

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs
```

```python
bank_data = make_blobs(n_features=10000, n_samples=5000, centers=4,  cluster_std=40, random_state=2018)
print(bank_data[0].shape)
```

```python
pca = PCA(n_components=2, svd_solver="auto")
results = pca.fit_transform(bank_data[0])
```

```python
plt.scatter(results[:, 0], results[:, 1], c=bank_data[1], alpha=0.3)
```
